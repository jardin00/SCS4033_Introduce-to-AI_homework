```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# Feature 1개인 경우
## 기업 매출 예측
- feature: 광고 비용(Ad)
- output: 기업 매출 (Sales)

```python
Ad = 1000 * np.random.rand(100, 1) #광고비용, 100개의 행과 1개의 열(column)을 가지는 2차원 NumPy 배열을 생성
Sales = 3 * Ad + 100*np.random.randn(100, 1) #기업 매출, 100개의 행과 1개의 열(column)을 가지는 2차원 NumPy 배열을 생성
```

```python
plt.plot(Ad, Sales, "b.")
plt.xlabel("$Ad$", fontsize=15)
plt.ylabel("$Sales$", rotation=0, fontsize=15)
plt.show()
```

```python
#예측값 만들기
Ad_new = 1000 * np.random.rand(100, 1)
```

```python
LR = LinearRegression()
LR.fit(Ad, Sales) #모델 생성
print("절편:", LR.intercept_)
print("계수:", LR.coef_)
```

```python
y_pred = LR.predict(Ad_new) #예측하기

#예측값 그래프로 표현하기
plt.plot(Ad_new, y_pred, "b.")
plt.xlabel("Ad new", fontsize=15)
plt.ylabel("Prdict Sales", fontsize=15)
plt.show()
```
